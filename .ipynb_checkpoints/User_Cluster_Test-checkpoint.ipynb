```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import actual user data (with limited information)
data = pd.read_csv("user_profiles.csv", 
                    names = ["user_id", "last_name", "first_name", "role", "email", "date_created", "header"]) \
         .drop(29) \
         .drop(["user_id", "role", "date_created", "header"], axis = 1)
data = pd.concat([data, data]).reset_index().drop("index", axis = 1)

#generate simulated data
dummy = np.zeros(data.shape[0])

#time zone
data["time_zone"] = dummy
def generate_time_zones(x):
    time_zones = {"PST": 0.166, "MST": 0.233, "CST": 0.524, "EST": 1.0} #https://i.redd.it/rvvwwhggop111.jpg
    p = np.random.rand()
    for key in time_zones.keys():
        if p < time_zones[key]:
            return key
data["time_zone"] = data["time_zone"].apply(generate_time_zones)

#gender
data["gender"] = dummy
def generate_gender(x):
    genders = {"female": 0.75, "male": 1.0} #https://www.caregiver.org/resource/caregiver-statistics-demographics/
    p = np.random.rand()
    for key in genders.keys():
        if p < genders[key]:
            return key
data["gender"] = data["gender"].apply(generate_gender)

#age
data["age"] = dummy
def generate_age(x):
    mean = 49.2 #https://www.caregiver.org/resource/caregiver-statistics-demographics/
    std = (49.2 - 18) / 2.5
    return np.random.normal(mean, std)
data["age"] = data["age"].apply(generate_age)

#hours per week
def generate_hours_per_week(data):
    hours = {"14.8": 24, "19.3": 44, "25.8": 54, "25.3": 64, "30.7": 74, "34.5": np.inf} 
    #https://www.caregiver.org/resource/caregiver-statistics-demographics/
    lst = []
    for i in range(data.shape[0]):
        age = data["age"].iloc[i]
        for key in hours.keys():
            if age < hours[key]:
                mean = float(key)
                break
        lst.append(np.random.normal(mean, 3))
    data["hours_per_week"] = lst
generate_hours_per_week(data)

#years of experience
def generate_years_of_experience(data):
    lst = []
    mean_age = np.mean(data["age"])
    std_age = np.std(data["age"])
    for i in range(data.shape[0]):
        z_age = (data["age"].iloc[i] - mean_age) / std_age
        val = (z_age + 4) + 4 + np.random.normal(0, 1)
        if val < 1:
            lst.append(np.random.normal(1, 0.3))
        else:
            lst.append(val)
    data["years_of_experience"] = lst
generate_years_of_experience(data)

from scipy.stats import percentileofscore
def generate_workload_index(data):
    lst = []
    for i in range(data.shape[0]):
        percentile_hours = percentileofscore(data["hours_per_week"], data["hours_per_week"].iloc[i])
        percentile_exp = percentileofscore(data["years_of_experience"], data["years_of_experience"].iloc[i])
        lst.append((percentile_hours + percentile_exp) / 2)
    data["workload_index"] = lst
generate_workload_index(data)
        
#relationship with care receiver
def generate_relationship_with_receiver(data):
    relationships = {"mother": 0.31 / 0.85, "father": 0.42 / 0.85, "non-relative": 0.57 / 0.85, "child": 0.71 / 0.85, 
                     "parent-in-law": 0.78 / 0.85, "grandparent": np.inf}
    #https://www.caregiver.org/resource/caregiver-statistics-demographics/
    lst = []
    for i in range(data.shape[0]):
        gender = data["gender"].iloc[i]
        if gender == "male": 
            val = np.random.normal(0.4, 0.2)
        else:
            val = np.random.normal(0.6, 0.2)
        for key in relationships.keys():
            if val < relationships[key]:
                lst.append(key)
                break
    data["relationship_with_receiver"] = lst
generate_relationship_with_receiver(data)

#residence
def generate_residence_with(data):
    lst = []
    residence = {"alone": 0.43 / 0.84, "partner": 0.7 / 0.84, "child": np.inf}
    gender = pd.get_dummies(data[["gender"]], drop_first = True).iloc[:, 0]
    mean_gender = np.mean(gender)
    std_gender = np.std(gender)
    for i in range(data.shape[0]):
        z_gender = (gender.iloc[i] - mean_gender) / std_gender
        val = np.random.rand() + z_gender / 5
        for key in residence.keys():
            if val < residence[key]:
                lst.append(key)
                break
    data["residence_with"] = lst
generate_residence_with(data)

#inputed first log
def generate_has_first_log(data):
    lst = []
    mean_age = np.mean(data["age"])
    std_age = np.std(data["age"])
    for i in range(data.shape[0]):
        z_age = (data["age"].iloc[i] - mean_age) / std_age
        val = np.random.rand() + (z_age / 4)
        if val < 0.5:
            lst.append(1)
        else:
            lst.append(0)
    data["has_first_log"] = lst
generate_has_first_log(data)
    
data.head()
```

```python
#encode data using one hot encoding
def get_enconded_data(data):
    data = data.copy()
    for col in data.columns:
        if type(data[col].iloc[0]) == str:
            ohe = pd.get_dummies(data[[col]], drop_first = False)
            data = data.drop(col, axis = 1)
            data = pd.merge(left = data, right = ohe, left_index = True, right_index = True)
    data = (data - data.apply(np.mean, axis = 0)) / data.apply(np.std, axis = 0)
    return data
data_encoded = get_enconded_data(data.iloc[:, 4:])

#perform SVD 
u, s, vt = np.linalg.svd(data_encoded, full_matrices = False)
usig = u * s
pc1 = usig[:, 0]
pc2 = usig[:, 1]
pc3 = usig[:, 2]
```

```python
#PCA pt.2
import plotly.graph_objs as go
data_merged = data.copy()
data_merged["pc1"] = pc1
data_merged["pc2"] = pc2
data_merged["pc3"] = pc3
def get_rgb(data):
    colors = []
    for i in range(data.shape[0]):
        if data["gender"].iloc[i] == "male":
            rgb = np.array([0, 0, 255.0])
        else:
            rgb = np.array([255.0, 0, 0])
        percentile_age = percentileofscore(data["age"], data["age"].iloc[i])
        rgb *= 1 - percentile_age / 150
        rgb = rgb.astype(int)
        colors.append(tuple(rgb))
    data["color"] = colors
get_rgb(data_merged)

trace = go.Scatter3d(x = data_merged["pc1"], y = data_merged["pc2"], z = data_merged["pc3"], 
                     mode='markers', 
                     hovertext = data_merged["first_name"] + [" "] * data_merged.shape[0] + data_merged["last_name"],
                     marker = {
                         "size": 8,
                         "color": [f'rgb({rgb[0]},{rgb[1]},{rgb[2]})' for rgb in data_merged["color"]]
                     })
fig = go.Figure(data = [trace], layout = go.Layout(margin = {"l": 0,"r": 0,"b": 0, "t": 0}))
fig.show()
```
